```python
cat_feats = ['purpose']
final_data = pd.get_dummies(loans, columns=cat_feats, drop_first=True)
```

```python
final_data.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = final_data.drop('not.fully.paid', axis=1)
y = final_data['not.fully.paid']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.30, random_state=101
)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier()
dtree.fit(X_train, y_train)
```

```python
predictions = dtree.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test, predictions))
```

```python
print(confusion_matrix(y_test, predictions))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators=600)
rfc.fit(X_train, y_train)
```

```python
rfc_pred = rfc.predict(X_test)
```

```python
print(classification_report(y_test, rfc_pred))
```

```python
print(confusion_matrix(y_test, rfc_pred))
```

The Random Forest model performed better in terms of overall accuracy. It achieved an accuracy of 85%, while the Decision Tree achieved 73%. This is because Random Forest combines many decision trees, which usually makes the model more stable and reduces overfitting.

However, the Decision Tree had better recall for class 1 (not fully paid), meaning it was better at detecting borrowers who did not fully pay. The Random Forest predicted class 0 very well, but it struggled to correctly identify class 1 because the dataset is imbalanced.

Overall, Random Forest performed better based on accuracy, but Decision Tree was better at identifying the minority class.
